## Уровень 0

```python
import sympy
from scipy.misc import derivative
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = sympy.symbols('x')
exp = x ** 5 + 4 * sympy.sin(2 * x) + sympy.cos(3 * x + 3)
f = lambda x: x ** 5 + 4 * np.sin(2 * x) + np.cos(3 * x + 3)
dif_1 = sympy.diff(exp, x)
print("\033[1mFirst derivative expression\033[0m:\n{}".format(dif_1))
dif_1_value = derivative(f, 1, n=1)
print("\033[1mFirst derivative value, x = 1\033[0m:\n{}".format(dif_1_value))
dif_2 = sympy.diff(dif_1, x)
print("\033[1mSecond derivative expression\033[0m:\n{}".format(dif_2))
dif_2_value = derivative(f, 1, n=2)
print("\033[1mSecond derivative value, x = 1\033[0m:\n{}".format(dif_2_value))
```

## Уровень 1

```python
f = lambda x: np.sin(2 * x + 1) ** 5

x = np.linspace(-5, 5)
fx = f(x)
f1x = [derivative(f, xx, dx=1e-6, n=1) for xx in x]
f2x = [derivative(f, xx, dx=1e-6, n=2) for xx in x]

plt.plot(x, f1x, label='1st derivative')
plt.plot(x, f2x, label='2nd derivative')
plt.legend()
plt.show()
```
