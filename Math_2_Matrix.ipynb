## Уровень 1

```python
import numpy as np
import matplotlib.pyplot as plt

A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)
x = A[:,0]
y = A[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

### Уменьшить объект в два раза

```python
half_mtrx = np.array([[0.5, 0, 0],
                      [0, 0.5, 0],
                      [0, 0, 0.5]])
A_half = A @ half_mtrx
```

```python
x = A_half[:,0]
y = A_half[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

### Повернуть на 130 градусов

```python
angle = np.deg2rad(130) 
turn_mtrx = np.array([[np.cos(angle), np.sin(angle),  0],
                      [-np.sin(angle), np.cos(angle), 0],
                      [0, 0, 1]])
```

```python
B = A @ turn_mtrx
```

```python
x = B[:,0]
y = B[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

### Отразить объект относительно прямой y=x

```python
mirror_mtrx = np.array([[0, 1, 0],
                        [1, 0, 0],
                        [0, 0, 1]])
```

```python
C = A @ mirror_mtrx
```

```python
x = C[:,0]
y = C[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

### Выполнить смещение объекта и масштабирование

```python
# 1. Выполняем смещение
bias_mtrx = np.array([[200, 300, 0] for _ in range(A.shape[0])])
A_bias = A + bias_mtrx

# 2. Выполняем масштабирование
scale_mtrx = np.array([[0.5, 0, 0],
                       [0, 1.2, 0],
                       [0, 0, 1]])
A_scale = A_bias @ scale_mtrx
```

```python
x = A_scale[:,0]
y = A_scale[:,1]

plt.plot(x, y)
plt.ylim([100, 500])
plt.xlim([0, 400])
```
