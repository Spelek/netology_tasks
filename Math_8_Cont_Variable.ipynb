```python
import scipy.stats as sps
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Нормальное распределение

```python
grid = np.linspace(-9, 10, 2000)  # сетка для построения графика
loc_values = [-5, -5, 0, 5, 5]  # набор значений параметра mu
sigma_values = [1, 2, 1, 2, 1]  # набор значений параметра sigma

plt.figure(figsize=(12, 6))

for i, (a, sigma) in enumerate(zip(loc_values, sigma_values)):
    plt.plot(grid, sps.norm(a, sigma).pdf(grid), lw=5, 
             label='$\mathcal{N}' + '({}, {})$'.format(a, sigma))

plt.legend(fontsize=16, loc='upper right')
plt.title('Плотности нормального распределения', fontsize=20)
plt.xlabel('Значение', fontsize=18)
plt.ylabel('Плотность', fontsize=18)
plt.grid(ls=':')
plt.show()
```

**Выводы:** параметр mu отвечает за сдвиг графика по оси X, а параметр sigma за сжатие/растяжение по обеим осям.

 Экспоненциальное распределение

```python
grid = np.linspace(-12, 20, 2000)  # сетка для построения графика
lam_values = [-10, -5, 0.1, 5, 10]  # набор значений параметра lam

plt.figure(figsize=(12, 6))

for lam in lam_values:
    plt.plot(grid, sps.expon(lam).pdf(grid), lw=5, 
             label=f'lam = {lam}')

plt.legend(fontsize=16, loc='upper right')
plt.title('Плотности экспоненциального распределения', fontsize=20)
plt.xlabel('Значение', fontsize=18)
plt.ylabel('Плотность', fontsize=18)
plt.grid(ls=':')
plt.show()
```

**Выводы:** параметр lam отвечает за сдвиг графика по оси X.

Распределение Стьюдента

```python
grid = np.linspace(-12, 20, 1000)  # сетка для построения графика
x_values = [1, 3, 3, 30, 30]
q_values = [0, 2, 5, 0, 0]
df_values = [1, 3, 3, 6, 2]

plt.figure(figsize=(12, 6))

for x, q, df in zip(x_values, q_values, df_values):
    plt.plot(grid, sps.t(x, q, df).pdf(grid), lw=5, 
             label=f'x, q, df = {x}, {q}, {df}')

plt.legend(fontsize=16, loc='upper right')
plt.title('Плотности распределения Стьюдента', fontsize=20)
plt.xlabel('Значение', fontsize=18)
plt.ylabel('Плотность', fontsize=18)
plt.grid(ls=':')
plt.show()
```

**Выводы:** параметры *x* и *df* отвечают за сжатие/растяжение по обеим осям, а параметр *q* за сдвиг по оси X.

Первый случай

```python
import math

def shannon(*args):
    res = 0
    total = sum(args)

    for arg in args:
        p = arg / total
        res -= p * math.log2(p)
    
    return res
```

```python
red = 54
yellow = 4
green = 28

print(shannon(red, yellow, green))
```
