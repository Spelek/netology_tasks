### Задание 1

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1., 1.])
y = np.array([2., 0.])
z = np.array([0., 2.])
```

```python
plt.figure()
plt.plot(*x, 'o')
plt.plot(*y, 'o')
plt.plot(*z, 'o')
plt.grid()
```

### Задание 2

```python
plt.figure()
plt.plot(*(x+y+z), 'o')
plt.grid()
```

### Задание 3

```python
cos_ = np.dot(x,y) / (np.linalg.norm(x) * np.linalg.norm(y))
angle = np.arccos(np.clip(cos_, -1, 1))  # Угол в радианах
angle_deg = np.rad2deg(angle)  # Угол в градусах
angle_deg
```
