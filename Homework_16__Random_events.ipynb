# Случайные события. Случайные величины

```
import numpy as np
```

## Моделирование игрового автомата 777
### Правила игры

```
# Награды за сочетания
rules = {
         1: {7: 3,
             0: 1},
         2: {7: 5,
             0: 2},
         3: {1: 10,
             3: 15,
             5: 50,
             7: 200,
             9: 100}
        }

# Определение выигрыша
def player_try(bingo=rules):
    # Ход игрока
    num1, num2, num3 = np.random.randint(0, 10, 3)

    # Определяем выигрыш
    if num1 == num2 == num3 and num1 in (1, 3, 5, 7, 9):
        return bingo[3][num1]
    elif num2 == num3 and num2 in (0, 7):
        return bingo[2][num2]
    elif num3 in (0, 7):
        return bingo[1][num3]
    else:
        return 0
```

### Игровой процесс

```
total = 0
N = 1000000

for _ in range(N):
    prize = player_try()
    if prize:
        total += prize
    # Учитываем цену игры
    total -= 1

print(f'Выигрыш за {N} попыток:', total, 'Средний выигрыш:', total/N)
```

### Вычисление среднего выигрыша игрока:
P(777|999|555|333|111) = 1/1000 = 0.001  
P(\*77) = 9/1000 = 0.009  
P(\**7) = 10\*10/1000 - 1/1000 - 9/1000 = 0.1 - 0.01 = 0.09  
P(\*00) = 10/1000 = 0.01  
P(\**0) = 10\*10/1000 - 10/1000 = 0.1 - 0.01 = 0.09  
Средний выигрыш = 0.001\*(200+100+50+15+10) + 0.009\*5 + 0.09\*3 + 0.01\*2 + 0.09\*1 = 0.375 + 0.045 + 0.27 + 0.02 + 0.09 = 0.42 + 0.38 = 0.8 руб.  
Учитывая цену игры в 1 руб., средний выигрыш = 0.8 - 1 = -0.2 руб.  
То есть в среднем игрок будет проигрывать (примерно 0.2 рубля) за каждую попытку(игру).  
Игра явно невыгодна игроку.
